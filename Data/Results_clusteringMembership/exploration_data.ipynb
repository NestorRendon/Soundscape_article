```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.stats.stattools import durbin_watson

# Load the CSV file into a DataFrame

# Handle missing values if any
df.dropna(inplace=True)

# Select relevant columns
selected_columns = ['component3_prob', 'ICHN200_b200mean_200', 'NDSI']
df = df[selected_columns]

# Plot a scatter plot to visualize the relationship between NDSI and ICHN200_b200mean_200
plt.scatter(df['NDSI'], df['ICHN200_b200mean_200'])
plt.xlabel('NDSI')
plt.ylabel('ICHN200_b200mean_200')
plt.title('Scatter Plot of NDSI vs. ICHN200_b200mean_200')
plt.show()

# Fit a simple linear regression model
X = df['NDSI']
y = df['ICHN200_b200mean_200']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

# Check for autocorrelation in residuals
dw_statistic = durbin_watson(model.resid)
print("Durbin-Watson statistic:", dw_statistic)

# Plot residuals vs. fitted values
residuals = model.resid
fitted_values = model.fittedvalues
plt.scatter(fitted_values, residuals)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Fitted Values')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()

# Plot histogram of residuals
plt.hist(residuals, bins=20, edgecolor='black')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()

# Fit the GLM
X = df['NDSI']
y = df['component3_prob']
X = sm.add_constant(X)
model = sm.GLM(y, X, family=sm.families.Binomial()).fit()

# Print model summary
print(model.summary())

# Plot the results
plt.scatter(df['NDSI'], df['component3_prob'], label='Data')
plt.plot(df['NDSI'], model.predict(), color='red', label='GLM Prediction')
plt.xlabel('NDSI')
plt.ylabel('Component3 Probability')
plt.title('GLM Prediction vs. Actual Data')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the CSV file into a DataFrame
df = pd.read_csv("/content/drive/MyDrive/Doctorado/2024/Todoslosindices08.csv")

# Handle missing values if any
df.dropna(inplace=True)

# Select relevant columns
selected_columns = ['component3_prob', 'ICHN200_b200mean_200', 'NDSI']
df = df[selected_columns]

# Round the values of component3_prob to integers



df['component3_prob'] = df['component3_prob'].round().astype(int)
# Split the data into two subsets based on component3_prob values (0 and 1)
df_0 = df[df['component3_prob'] == 0]
df_1 = df[df['component3_prob'] == 1]

# Fit GLM for component3_prob = 0
X_0 = df_0['NDSI']
y_0 = df_0['ICHN200_b200mean_200']
X_0 = sm.add_constant(X_0)
model_0 = sm.GLM(y_0, X_0, family=sm.families.Binomial()).fit()

# Fit GLM for component3_prob = 1
X_1 = df_1['NDSI']
y_1 = df_1['ICHN200_b200mean_200']
X_1 = sm.add_constant(X_1)
model_1 = sm.GLM(y_1, X_1, family=sm.families.Binomial()).fit()

# Print model summaries
print("Model for component3_prob = 0:")
print(model_0.summary())
print("\nModel for component3_prob = 1:")
print(model_1.summary())

# Plot the results for component3_prob = 0
plt.figure(figsize=(10, 5))
plt.scatter(df_0['NDSI'], df_0['ICHN200_b200mean_200'], label='Component3_Prob = 0')
plt.plot(df_0['NDSI'], model_0.predict(), color='red', label='GLM Prediction (Component3_Prob = 0)')
plt.xlabel('NDSI')
plt.ylabel('ICHN200_b200mean_200')
plt.title('GLM Prediction vs. Actual Data (Component3_Prob = 0)')
plt.legend()
plt.show()

# Plot the results for component3_prob = 1
plt.figure(figsize=(10, 5))
plt.scatter(df_1['NDSI'], df_1['ICHN200_b200mean_200'], label='Component3_Prob = 1')
plt.plot(df_1['NDSI'], model_1.predict(), color='green', label='GLM Prediction (Component3_Prob = 1)')
plt.xlabel('NDSI')
plt.ylabel('ICHN200_b200mean_200')
plt.title('GLM Prediction vs. Actual Data (Component3_Prob = 1)')
plt.legend()
plt.show()
```

```python
# Deviance Goodness-of-Fit Test
deviance_0 = model_0.deviance
deviance_1 = model_1.deviance

# Degrees of freedom for the models
df_model_0 = model_0.df_model
df_model_1 = model_1.df_model

# Chi-square critical value for significance level 0.05 and the corresponding degrees of freedom
chi2_critical = sm.stats.chi2.ppf(0.95, df_model_0)
print("Deviance Goodness-of-Fit Test for component3_prob = 0:")
print(f"Deviance: {deviance_0}")
print(f"Degrees of freedom: {df_model_0}")
print(f"Chi-square critical value: {chi2_critical}")
print(f"Deviance is {'significant' if deviance_0 > chi2_critical else 'not significant'} at the 0.05 level.\n")

chi2_critical = sm.stats.chi2.ppf(0.95, df_model_1)
print("Deviance Goodness-of-Fit Test for component3_prob = 1:")
print(f"Deviance: {deviance_1}")
print(f"Degrees of freedom: {df_model_1}")
print(f"Chi-square critical value: {chi2_critical}")
print(f"Deviance is {'significant' if deviance_1 > chi2_critical else 'not significant'} at the 0.05 level.\n")

# Hosmer-Lemeshow Test
print("Hosmer-Lemeshow Test:")
print("This test is not directly available in the statsmodels library. You may need to implement it manually or use an external package.")

```

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Leer el archivo CSV en un DataFrame
df = pd.read_csv("/content/drive/MyDrive/Doctorado/2024/Todoslosindices08.csv")

# Ajustar el GLMM con NDSI como variable de respuesta y component3_prob y ICHN200_b200mean_200 como variables predictoras
model = smf.mixedlm("NDSI ~ component3_prob + ICHN200_b200mean_200", df, groups=df["component3_prob"]).fit()

# Obtener los valores predichos
predicted_values = model.predict()

# Separar los valores predichos segn si son redondeados a 0 o 1 de la variable predictora ICHN200_b200mean_200
df_round_0 = df[df["ICHN200_b200mean_200"].round() == 0]
df_round_1 = df[df["ICHN200_b200mean_200"].round() == 1]

predicted_values_round_0 = predicted_values[df_round_0.index]
predicted_values_round_1 = predicted_values[df_round_1.index]

# Graficar los valores predichos separados por redondeo de la variable predictora
plt.figure(figsize=(8, 6))
plt.scatter(df_round_0["NDSI"], predicted_values_round_0, alpha=0.5, label="Predicciones para round 0")
plt.scatter(df_round_1["NDSI"], predicted_values_round_1, alpha=0.5, label="Predicciones para round 1")
plt.plot(df["NDSI"], df["NDSI"], color='red', label="Lnea del modelo lineal", linestyle='--')
plt.xlabel("NDSI observado")
plt.ylabel("NDSI predicho")
plt.title("Valores predichos vs. observados")
plt.legend()
plt.show()
```

```python
import statsmodels.stats.api as sms

# Obtener los residuos del modelo
residuals = model.resid

# Prueba de normalidad de los residuos (prueba de Shapiro-Wilk)
shapiro_test = sms.normal_ad(residuals)
print("Prueba de normalidad de los residuos (Shapiro-Wilk):")
print("Estadstica de prueba:", shapiro_test[0])
print("p-valor:", shapiro_test[1])

# Prueba de homocedasticidad (prueba de Breusch-Pagan)
bp_test = sms.het_breuschpagan(residuals, model.model.exog)
print("\nPrueba de homocedasticidad (Breusch-Pagan):")
print("Estadstica de prueba:", bp_test[0])
print("p-valor:", bp_test[1])

# Prueba de independencia de los residuos (prueba de Durbin-Watson)
dw_test = sms.durbin_watson(residuals)
print("\nPrueba de independencia de los residuos (Durbin-Watson):")
print("Estadstica de prueba:", dw_test)


```

```python
# Prueba de linealidad del modelo
linear_test = sms.linear_harvey_collier(model)
print("\nPrueba de linealidad del modelo (Harvey-Collier):")
print("Estadstica de prueba:", linear_test[0])
print("p-valor:", linear_test[1])
```
